## 3.1 Basic Plotting

SymPy has plotting features built-in, so no need to import the Matplotlib package if you only intend to plot symbolic functions.

**SymPy plots vs. Matplotlib?** SymPy's plotting feature is a *fork* of Matplotlib, a modified version. You don't need to import Matplotlib itself if you are only plotting SymPy expressions.

Plotting Overview:
- If you liked MATLAB's style of plotting, you'll like this.
- *State-based*: commands are directives to change the state of the *most recently created plot*

To plot a single-variable function, use the **sp.plot** command:

**sp.plot(function, (x, leftbd, rightbd), optional_additional_arguments)**

There are MANY additional arguments you can include to customize your plot! For a complete list, see https://docs.sympy.org/latest/modules/plotting.html

Here is an example plot:

```python
x = sp.symbols('x')
f = -x**2 + 9

fig1 = sp.plot(f, (x,-3,3), line_color = 'red')
```

## 3.2 Multiple Function Plots

If we have more than one function to graph, we have to:
- Create the first plot, but include the argument **show = False** to tell it to not display just yet.
- Create the second plot, also with **show = False**
- Use Python's **.extend()** command to dump the contents of one plot into the other. (Plots are lists at heart!)
- Use the **.show()** command to show the plot:

```python
plot1 = sp.plot(f, (x,-3,3), line_color = 'red', show = False)
plot2 = sp.plot(9, (x,-3,3), line_color = 'blue', show = False)
plot1.extend(plot2)
plot1.show()
```

## 3.3 Customization

Here is an example of the first plot in the notebook with some customizations added:

```python
fig2 = sp.plot(f,                            # Each argument given to sp.plot can be separated by a newline for
               (x,-3,3),                     # readability, but you don't have to.
               line_color = 'red',
               xlim = (-5,5),
               ylim = (-1,10), 
               title='Plot of $9 - x^2$')    # YES, SymPy plots allow LaTeX math mode for titles!!!
```

**Your turn:** Below is a (not so comprehensive) list of all the input arguments sp.plot will take for plot customization. Try a few on the plot above, or if you're bored with the parabola example, define a new function to plot and make it look nice.

### Customization Options:

- **line_color**: can take a string (example, **line_color = 'cyan'**) or a number (see the link below for info on color numbering)
- **title**: takes a string to place at the top.
- **xlabel** and **ylabel**: takes a string to use as an axis label.
- **xscale** and **yscale**: takes 'linear' or 'log' as inputs. Default is linear.
- **axis_center**: takes a point $(x,y)$ to set as the center of the plot.
- **xlim** and **ylim**: takes intervals $(min,max)$ as ranges for $x$ and $y$.
- **size**: takes dimensions $(width,height)$ in inches to resize figure.

To see all possible options for **sp.plot**, see https://docs.sympy.org/latest/modules/plotting.html#sympy.plotting.plot.plot

### Other plotting commands:

Similar to the above, each link will take you to the corresponding command's syntax and full list of customization options.

- **plot_parametric()** https://docs.sympy.org/latest/modules/plotting.html#sympy.plotting.plot.plot_parametric
- **plot_implicit()** https://docs.sympy.org/latest/modules/plotting.html#sympy.plotting.plot_implicit.plot_implicit
- **plot3d()** https://docs.sympy.org/latest/modules/plotting.html#sympy.plotting.plot.plot3d
- **plot3d_parametric_line()** https://docs.sympy.org/latest/modules/plotting.html#sympy.plotting.plot.plot3d_parametric_line
- **plot3d_parametric_surface()** https://docs.sympy.org/latest/modules/plotting.html#sympy.plotting.plot.plot3d_parametric_surface

### Bonus: tikzplotlib

This package can take the most recently displayed plot and convert it to TikZ code for you! First, install it using the terminal:
- Close out of Jupyter Notebook by File > Close and Halt.
- Go back to the terminal and type **pip install tikzplotlib** (we can talk about using **pip install** vs. **conda install** at a later point).
- Re-enter the notebook and it's ready to import!

```python
import tikzplotlib

tikzplotlib.clean_figure() 
tikzplotlib.save('piecewise.tex')
```

**clean_figure()** trims anything not shown in the current display window (such as functions drawn for a larger domain than what is shown) to prepare it for conversion.

**save(filename.tex)** is self-explanatory. The file is in the same directory as the notebook.
